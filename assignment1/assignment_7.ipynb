## 2. 
- Compute the expected value for a uniform random variable.
- Show that $\mathbb{E}[a+bX] = a + b\mathbb{E}[X]$
- Show, by example, that $v(\mathbb{E}[X]) \neq \mathbb{E}[v(X)]$, if $v(x) \neq a+bx$. For example, try $v(y) = y^2$ or $v(y)=\sqrt{y}$ with a Bernoulli or uniform or normally distributed random variable. This can be an important thing to remember: The expectation of a transformed random variable is not the transformation of the expected value.

```python
def v(x):
    return np.sqrt(x)
s = np.random.uniform(low=0, high=1,size = 20000)

eos = np.mean( v(s) )
print(f'Expected Value: {np.round(eos, 3)}')

vex = v(np.mean(s))
print(f'Value of Expectation: {np.round(vex, 3)}')

sns.lineplot(x=s, y=v(s), color='green', label='v(x)')
plt.plot(np.mean(s), eos, 'o', color='purple', markersize=10)
```

```python
path = r"C:\Users\crush\OneDrive\Desktop\DS 5030\assignment1\data\metabric.csv"
df = pd.read_csv(path)
y = df['Radio Therapy']
sns.histplot(y,stat='proportion', color='red')
```

```python
y = pd.to_numeric(y.replace({'NO':'0','YES':'1'}), errors='coerce')
print(f'Maximum Likelihood Estimate:', np.round(y.mean(), 2))

s = 2500
sample = [np.mean(y.sample(frac=1,replace=True)) for s in range(s)]

# plot kde
fig, axes = plt.subplots(1, 2, figsize=(8,4))
sns.kdeplot(x=sample, ax=axes[0], label='kde', color='orange')

# plot ecdf
axes[0].set_title('Sample KDE')
sns.ecdfplot(x=sample, ax=axes[1], label='ecdf', color = "green")

axes[1].set_title('Sample ECDF')
plt.show()
```
